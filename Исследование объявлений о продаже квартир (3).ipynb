## Предобработка данных

Подсчитаем процентное количество пропусков в нашей таблице при помощи метода data.isna().mean()

```python
#считаем процентное количество пропусков в нашей таблице
pd.DataFrame(round((data.isna().mean()*100),2)).style.background_gradient('coolwarm')
```

Исходя из полученных данных, выберем столбцы, в которых пропуски не превышают 10 % и заполним их  медианным значением. Это столбцы `living_area` и `kitchen_area`. В столбце `locality_name` приведем названия населенных пунктов к нижнему регистру и заменим ё на е.

В столбце `is_apartment` нам следует поменять тип данных с object на bool (что указано в вводных данных). Пропуски в столбце `balcony` мы можем заменить на 0, поскольку, скорее всего, раз количество балконов арендадателем не указано, то балконов вообще нет.

В столбце `days_exposition` пропуски мы заменим разностью из максимального значения столбца `first_day_exposition` и значением этого же столбца с пропуском в столбце `days_exposition`. 

Пропуски столбцов `airports_nearest`,`cityCenters_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest` мы заменим на значение '0'.

Пропуски в `ceiling_height` мы трогать не будем, чтоб не исказить полученные данные. Количество же пропусков в остальных столбцах  довольно мало, поэтому мы можем их опустить. 


Начнем с самого простого: заменим пропущенные значения в столбце `balcony` на '0', используя метод fillna()

```python
#заполняем пропуски в столбце 'balcony' значением равному нулю при помощи метода fillna()
data['balcony'] = data['balcony'].fillna('0')
```

```python
#проверяем колличество пропусков в столбце при помощи метода isna().sum()
data['balcony'].isna().sum()
```

Заполнив значения в `balcony`, мы переходим к столбцу `ceiling_height` и остальным столбцам с пропусками. Напишем цикл, при котором пропущенные значения будут заменены медианным значением в зависимости от локации.

Затем будем заполнять пропуски в каждом столбце медианным значением.

```python
# находим изначальное количество пропусков в столбце 'living_area'
print('Пропуски до:', data['living_area'].isna().sum())

for name in data['locality_name'].unique():
    median = data.loc[data['locality_name'] == name, 'living_area'].median()
    data.loc[(data['locality_name'] == name)&(data['living_area'].isna()), 'living_area'] = median
    
# находим итоговое количество пропусков в столбце 'living_area' после замены пропусков
print('Пропуски после:', data['living_area'].isna().sum())
```

```python
# находим изначальное количество пропусков в столбце 'kitchen_area'
print('Пропуски до:', data['kitchen_area'].isna().sum())

for name in data['locality_name'].unique():
    median = data.loc[data['locality_name'] == name, 'kitchen_area'].median()
    data.loc[(data['locality_name'] == name)&(data['kitchen_area'].isna()), 'kitchen_area'] = median
    
# находим итоговое количество пропусков в столбце 'kitchen_area' после замены пропусков
print('Пропуски после:', data['kitchen_area'].isna().sum())
```

Теперь приведем названия населенны пунктов к нижнему регистру и заменим ё на е.

```python
# выполним перевод в нижний регистр всех значений в столбце 'locality_name'
data['locality_name'] = data['locality_name'].str.lower()
```

```python
# выполним замену 'ё' на 'е' всех значений в столбце 'locality_name'
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
```

Далее заполняем пропущенные значение в столбце `days_exposition`. Для этого приведем столбец `first_day_exposition` из строки в дату и вычтем из его максимального значения значение этого же столбца с пропуским в столбце `days_exposition`.

```python
# используем метод to_datetime() для перевода строки в дату
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])
# при помощи loc() выделяем максимальное значеие столбца 
# и вычитаем из него пустое значение столбца days_exposition
maximum = data['first_day_exposition'].max()
data.loc[data['days_exposition'].isna(),'days_exposition'] = (
    (maximum-data.loc[data['days_exposition'].isna(), 'first_day_exposition'])
    .dt.days
)
```

Также, остались пропуски в столбце `is_appartment`. По условию, значения этого столбца носят тип bool, на который мы и поменяем его.

```python
# приводим значение столбца is_apartment к формату bool, используя astype()
data['is_apartment'] = data['is_apartment'].astype('bool')
```

В оставшихся столбцах, в которых указан километраж, заполним пропуски значением '0' и изменим тип данных на целочисленное.

```python
# используем метод fillna() для заполнения пропусков
# и метод astype(), чтоб заменить значение столбца на целочисленное
data['airports_nearest'] = data['airports_nearest'].fillna('0').astype('int')
data['cityCenters_nearest'] = data['cityCenters_nearest'].fillna('0').astype('int')
data['parks_around3000'] = data['parks_around3000'].fillna('0').astype('int')   
data['parks_nearest'] = data['parks_nearest'].fillna('0').astype('int')   
data['ponds_around3000'] = data['ponds_around3000'].fillna('0').astype('int')   
data['ponds_nearest'] = data['ponds_nearest'].fillna('0').astype('int')
```

```python
# проверяем количество пропусков в столбце при помощи метода isna().sum()
data.isna().sum()
```

**ВЫВОД**

Замена пропусков ниже 10 % была успешно осуществлена при помощи внедрения среднего значения по соответствующим показателям. Столбец `ceiling_height` мы не трогали, поскольку количество пропусков составляет около 38%, а замена их в такой ситуации на конкретное значение без ссылки на источники - это, скорее, искажение исходных данных, а не их обогащение.

Картографические данные (расстояние до парков и проч) наверняка заполнялись автоматически, поэтому сложно восстановить их каким-то известным способом. Мы заменили эти показатели на нули.

Названия населенных пунктов мы привели к нижнему регистру, произвести обработку буквы ё (замена на е). Это действие помогло нам избавленится от неявных дубликатов в названиях населенных пунктов.

## Расчёты и добавление результатов в таблицу

 В данной части проекта мы будем добавлять столбцы со значением
1. Цены квадратного метра;
2. Дня недели, месяц и год публикации объявления;
3. Этажа квартиры; варианты — первый, последний, другой;
4. Соотношения жилой и общей площади, а также отношение площади кухни к общей.

    1. Цена квадратного метра. 
Значение это находим путем разделения столбца с итоговой ценой - `last_price` на общую площадь помещения - `total_area`. Полученный результат сохраняем в столбец `price_per_square_meter`.

```python
# находим цену за квадратный метр и сохраняем в новый столбец price_per_square_meter
data['price_per_square_meter'] = data['last_price']/ data['total_area']
```

    2.День недели, месяц и год публикации объявления.
Приведем значение столбца `first_day_exposition` из строкового в дату. Затем округлим значения используя dt.weekday, dt.month и dt.year. Полученные результаты сохраняем в столбцы `weekday`, `month` и `year` соответственно.  

```python
# используем to_datetime() к столбцу 'first_day_exposition'
data['first_day_exposition'] = (
    pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
)
# округляем наши значения
data['weekday'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year
```

    3.Этаж квартиры; варианты — первый, последний, другой
Для этого столбца напишем функцию, которая будет считать этажи. Затем используем метод apply(). Итоговое значение сохраняем в столбец `floor_level`.

```python
# создаем функцию floor_level для подсчета этажей
def floor_level(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
# применяем фукцию к нашему датасету
data['floor_level'] = data.apply(floor_level, axis=1)
```

    4. Соотношение жилой и общей площади, а также отношение площади кухни к общей.
Значение это находим путем разделения столбцов `living_area` и `kitchen_area` на общую площадь помещения - `total_area`. Полученный результат сохраняем в столбцы `ratio_living_area` и `ratio_kitchen`.

```python
# подсчет соотношения жилой и общей площади, а также отношение площади кухни к общей.
data['ratio_living_area'] = data['living_area'] / data['total_area']
data['ratio_kitchen'] = data['kitchen_area'] / data['total_area']
```

 * цена квадратного метра - `price_per_square_meter`;

## Исследовательский анализ данных

На данном этапе мы изучим следующие параметры: 
- площадь;

- высота потолков.
   

Для изучения этих параметров напишем функцию, которая будет считать межквартильный размах для всех параметров.Узнаем межквартильный размах - ориентировочная граница нормального размаха. Всё, что лежит за ней, будем считать выбросом. Затем построим диаграмму размаха boxplot(), что узнать,какие значения оказались вдали от среднего, и насколько их много (характерный разброс). 
Затем построим гистограмму для каждого параметра.



Пишем функцию, которая будет считать нам межквартильный размах(IQR) - расстояние между первым(q1) и третьим квартилем(q3). В функции вызываем метод describe(), который покажет нам числовое описание данных - среднее, медиану, максимальное, минимальное, количество наблюдений в выбросе и их разброс. 25% - это q1(первая четверть выборки), 75% - это q3 (отсечка трех четвертей), в describe() они находятся в 4 и 6 колонке соответственно, a iqr это разница между ними. Затем находим минимальную и максимальную величину нашей выборки.

```python
# создаем функцию iqr_total(), которая посчитает iqr
def iqr_total(column):
    columns_describe = data[column].describe()
    
    # считаем iqr через разницу q3 и q1 
    iqr = columns_describe[6] - columns_describe[4]
    
    # считаем левую и правую границу iqr    
    left = columns_describe[4] - 1.5 * iqr
    right = columns_describe[6] + 1.5 * iqr

    # устанавилваем минимальное и масимльное значение для левой и правой границы
    if left < columns_describe[3]:
        left = columns_describe[3] 
    if right > columns_describe[7]:
        right = columns_describe[7]
        
    return [left, right]
```
